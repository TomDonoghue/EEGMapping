## How to fix event bullshitery.
- Fill out the ev_dict with all the events
    - Use a friendly mappin
        - 100X are starts
        - 200X are end
        - 30XX are trial type
- Load data, without events (set function to None)
- Use ev_dict in mne.io.eeglab.read_events_eeglab
    - Gives us that matrix of all events
- Subselect that matrix to [keep what we care about / drop what we don't care about]
- Add the evs matrix [n_evs, 3] shape, updated to have no overlaps, to eeg_dat
    - eeg_dat.add_events(evs)
